## Read {year,type,county} data extracted [by Chat GPT](https://chat.openai.com/share/cb4c2293-5320-47a2-8512-5e9ac823ac4d)

For years that are otherwise missing from the annual "Victim Classification by County" summary PDFs (see [NJSP summary PDFs.ipynb](../annual-summaries/NJSP%20summary%20PDFs.ipynb)).

```python
from utz import *
from njsp.paths import MISSING_YTC
```

```python
def read_annual_csv(year):
    a = read_csv(f'{year}.csv')
    a.columns = a.columns.str.lower()
    a = a.rename(columns={ 'pedalcyclist': 'cyclist', })
    a = a.set_index('county')
    a.index = a.index.str.capitalize()
    total = a.loc['Total']
    c = a.drop(index='Total')
    cols = [ 'driver', 'passenger', 'pedestrian', 'cyclist' ]
    assert (c.sum() == total).all()
    assert (a[cols].sum(1) == a.killed).all()
    return c.drop(columns='killed')
```

```python
def read_annual_csvs(years):
    df = pd.concat([
        read_annual_csv(year).assign(year=year)
        for year in years
    ]).reset_index()
    df['county'] = df['county'].str.replace('Cape may', 'Cape May')
    return df.set_index(['year', 'county'])
```

```python
c = read_annual_csvs([ 2008, 2009, 2017, 2018 ])
c
```

```python
c.to_csv(MISSING_YTC)
```
