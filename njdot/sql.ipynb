```python
from utz import *
from njdot import crashes, drivers, occupants, pedestrians, vehicles
from njdot.load import pk_base, INDEX_NAME, normalize, print_hists

from nj_crashes.utils import sql
import sqlite3
```

```python
%%time
d = drivers.load()
p = pedestrians.load()
c = crashes.load()
v = vehicles.load()
o = occupants.load()
```

```python
n = 1000000
db_path = f'all_{n}.db'
db_url = f'sqlite:///{db_path}'

def check(tbl, df1):
    df2 = read_sql(tbl, db_url, index_col='id')
    cols1 = set(df1.columns)
    cols2 = set(df2.columns)
    assert cols1 == cols2, f"{cols1.difference(cols2)} missing, {cols2.difference(cols1)} extra"
    df2 = df2[df1.columns]
    assert ((df1 == df2) | (df1.isna() & df2.isna())).all().all()
    return df2

nc = len(c)
cutoff = nc - n
c1 = c[c.index >= cutoff]
d1 = d[d.crash_id >= cutoff]
o1 = o[o.crash_id >= cutoff]
p1 = p[p.crash_id >= cutoff]
v1 = v[v.crash_id >= cutoff]

for tbl, df in {
    'c': c1,
    'd': d1,
    'o': o1,
    'p': p1,
    'v': v1,
}.items():
    df.to_sql(tbl, db_url, if_exists='replace')
    if tbl != 'c':
        with sqlite3.connect(db_path) as con:
            cur = con.cursor()
            sql.add_fk(cur, tbl, 'crash_id', 'c', 'id')
            if tbl == 'c':
                for idx_cols in [
                    ('severity', 'dt', 'cc', 'mc'),
#                     ('severity', 'ilat', 'ilon'),
#                     ('severity', 'icc', 'dt'),
                ]:
                    sql.add_idx(cur, 'c', *idx_cols)
        check(tbl, df)
```

```python
def condition_hist(df, name):
    cond = df.set_index('crash_id').condition.replace('', nan).dropna()
    cond = cond.loc[cond != 1]
    return cond.groupby(lambda x:x).apply(lambda s: ''.join(s.sort_values().astype(str))).rename(name)
```

```python
%%time
pih = condition_hist(p, 'pih')
pih
```

```python
def deaths_by_crash(df, name):
    cond = df.set_index('crash_id').condition.replace('', nan).dropna()
    cond = cond[cond == 1]
    return cond.groupby(lambda x:x).size().rename(name)
```

```python
%%time
pkn = deaths_by_crash(p, 'pkn')
pkn
```
