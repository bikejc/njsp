```python
from utz import *
from njdot import crashes, drivers, occupants, pedestrians, vehicles
from njdot.load import pk_base, INDEX_NAME, normalize, print_hists
```

```python
%%time
c = crashes.load()#write_pqt=True)
c
```

```python
vm = v.merge(c.year, how='left', left_on='crash_id', right_index=True)
vm
```

```python
pd.crosstab(
    ~v.cir1.isna().rename('has_cir1'),
    ~v.cir2.isna().rename('has_cir2'),
)
```

```python
has_ev  = ~v.ev .isna().rename('has_ev' )
has_ev1 = ~v.ev1.isna().rename('has_ev1')
has_ev2 = ~v.ev2.isna().rename('has_ev2')
has_ev3 = ~v.ev3.isna().rename('has_ev3')
has_ev4 = ~v.ev4.isna().rename('has_ev4')
has_evs = (has_ev1 | has_ev2 | has_ev3 | has_ev4 ).rename('has_evs')
evs = sxs(
    has_ev1,
    has_ev2,
    has_ev3,
    has_ev4,
    vm.year,
)
g = evs.groupby(['has_ev1', 'has_ev2', 'has_ev3', 'has_ev4'])
gy = g['year']
sxs(
    g.size().rename('num'),
    gy.min().rename('min'),
    gy.max().rename('max'),
)
```

```python
vev = vm[~vm.ev.isna()]
sxs(
    vev.ev == vev.ev1,
    vev.ev == vev.ev2,
    vev.ev == vev.ev3,
    vev.ev == vev.ev4,
).sum()
```

```python
%%time
vev.apply(lambda r: ' '.join(
    '   ' if isna(r[f'ev{i}']) else f'{i}✅' if r.ev == r[f'ev{i}'] else f'{i}❌'
    for i in range(1, 5)
), axis=1).value_counts().sort_index()
```
